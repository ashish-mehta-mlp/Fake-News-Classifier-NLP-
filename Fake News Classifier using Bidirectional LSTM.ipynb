```
from tensorflow.keras.layers import Embedding # Used to convert the sentence to vectors
from tensorflow.keras.preprocessing.sequence import pad_sequences # to make all sentences of equal size
from tensorflow.keras.preprocessing.text import one_hot # To convert words to numerical form
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dense
```

```
## Vocabulary size (when we are implementing One-Hot representation we need a dictionary size within which, all the words will be given some kind of sequence)

vocab_size = 5000
```

## One Hot Representation

```
message = X.copy()
```

```
message["title"][1]
```

```
message.reset_index(inplace = True)
```

```
import nltk
import re
from nltk.corpus import stopwords
```

```
nltk.download("stopwords")
```

```
# Dataset Preprocessing

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range (0, len(message)):
  print(i)
  review = re.sub("[^a-zA-Z]", " ", message["title"][i]) # remove all special characters from the title
  review = review.lower() # Converting all the words to lower case\
  review = review.split() # splitting each word

  review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
  review = " ".join(review)
  corpus.append(review)
```

```
corpus
```

```
# One hot encoding will assign some index to each word in the corpus

onehot_rep = [one_hot(words, vocab_size) for words in corpus]
```

```
onehot_rep
```

## Embedding Representation

- This is done to make all the sentences of the same size

```
sent_length = 25
embedded_doc = pad_sequences(onehot_rep, padding = "pre", maxlen = sent_length)
print(embedded_doc)
```

```
embedded_doc[0]
```

## Creating The Model

```
embedding_vector_features = 40 # each sentence in embedding_doc will be represented in 40 dimensions while embedding it to vectors
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_features, input_length=sent_length)) # embedding is used to convert the numerical values in embedded_doc to vector
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1, activation = "sigmoid"))
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
print(model.summary())
```

- When we use bidirectional LSTM, that time, two LSTM are created, one in forward direction and one in backward direction.

```
import numpy as np
X_final = np.array(embedded_doc)
y_final = np.array(y)
```

```
X_final.shape, y_final.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.25, random_state = 42)
```

## Model Training

```
model.fit(X_train, y_train, validation_data =(X_test, y_test), epochs = 10, batch_size = 64)
```

#### The final validation accuracy that we got is 90.73%

## Adding Dropout

```
from tensorflow.keras.layers import Dropout
embedding_vector_features = 40 
model1 = Sequential()
model1.add(Embedding(vocab_size, embedding_vector_features, input_length=sent_length))
model1.add(Dropout(0.3)) 
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.3)) 
model1.add(Dense(1, activation = "sigmoid"))
model1.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
print(model1.summary())
```

```
model1.fit(X_train, y_train, validation_data =(X_test, y_test), epochs = 10, batch_size = 64)
```

#### Even after adding dropout layers, the accuracy remains the same i.e. 90.73%

## Performance Metrics And Accuracy

```
y_pred = model.predict(X_test) # Without Dropout
y_pred1 = model1.predict(X_test) # With Dropout
```

```
from sklearn.metrics import confusion_matrix
```

```
#Without Dropout
confusion_matrix(y_test, y_pred.round())
```

```
#With Dropout
confusion_matrix(y_test, y_pred1.round())
```

```
#Without Dropout
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred.round())
```

```
#With Dropout
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred1.round())
```

## Precision, Recall and F1 Score

```
#Without Dropout
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred.round()))
```

```
#With Dropout
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1.round()))
```
